```python
df = spark.read.csv('dbfs:/databricks-datasets/rwe/ehr/csv/conditions.csv', header=True)

df.write.mode('overwrite').saveAsTable('demo_rwe_conditions')

print(df.schema)
display(df)
```

```python
# Create a UC Metric View for the clinical conditions dataset with table and column descriptions
spark.sql(
    """
CREATE OR REPLACE VIEW demo_rwe_conditions_metric_view
COMMENT 'Metric view for clinical conditions, including patient, encounter, and condition details for analytics.'
WITH METRICS
LANGUAGE YAML
AS $$
version: 1.1
source: demo_rwe_conditions

dimensions:
  - name: Start
    expr: START
    comment: The start date/time of the clinical condition.
  - name: Stop
    expr: STOP
    comment: The stop date/time of the clinical condition.
  - name: Patient
    expr: PATIENT
    comment: The patient identifier.
  - name: Encounter
    expr: ENCOUNTER
    comment: The encounter identifier.
  - name: Code
    expr: CODE
    comment: The code representing the clinical condition.
  - name: Description
    expr: DESCRIPTION
    comment: The description of the clinical condition.

measures:
  - name: Condition Count
    expr: COUNT(1)
    comment: The total number of condition records.
  - name: Distinct Patients
    expr: COUNT(DISTINCT PATIENT)
    comment: The number of unique patients with conditions.
$$
"""
)
```
